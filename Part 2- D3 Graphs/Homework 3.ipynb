````python
import pandas as pd
social_media = pd.read_csv('socialMedia.csv')
social_media.head()
````

````python
# find (and round) the avg number of likes, grouping by platform and post type
SocialMediaAvg = social_media.groupby(['Platform', 'PostType'])['Likes'].mean().reset_index()
SocialMediaAvg['Likes'] = round(SocialMediaAvg['Likes'], 2)
# rename to "AvgLikes"
SocialMediaAvg.rename(columns={'Likes': 'AvgLikes'}, inplace=True)
SocialMediaAvg.head()

# create new csv file for SocialMediaAvg
SocialMediaAvg.to_csv('SocialMediaAvg.csv', index=False)
````
